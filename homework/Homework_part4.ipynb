```python
import numpy as np
import matplotlib.pyplot as plt

#%matplotlib inline
```

```python
x_points = np.linspace(0, 10, 100)
y_points1 = np.array([x**2 - 6*x + 11 for x in x_points])
```

```python
y_points2 = np.array([6*x - 10 for x in x_points])
```

```python
y_points3 = np.array([np.exp(x/float(3)) for x in x_points])
```

```python
y_points4 = np.array([11 * np.sqrt(x) for x in x_points])
```

```python
plt.figure(figsize = (10, 7))
plt.plot(x_points, y_points1, color = "red", label = "$x^2 - 4x + 11$")
plt.plot(x_points, y_points2, color = "green", label = "$6x - 10$")
plt.plot(x_points, y_points3, color = "blue", label = "$e^{x/3}$")
plt.plot(x_points, y_points4, color = "orange", label = "$11\sqrt{x}$")
plt.grid()
plt.legend()
plt.xlim(0, 10)
plt.xlabel("Ось Х")
plt.ylabel("Ось Y")
plt.title("Графики функций")
plt.savefig("my_fig")
plt.show()
```

```python
x_points = np.linspace(-1, 3, 100)
y_points1 = np.array([x**2 for x in x_points])
y_points2 = np.array([x**3 - 2*x + 4 for x in x_points])
y_points3 = np.array([np.exp(x) for x in x_points])
y_points4 = np.array([np.log(x) if (x >= 0) else None for x in x_points])
```

```python
plt.figure(figsize = (10, 7))
type_of_graph = ["square", "cube", "exponent", "logarithm"]
y_points = [y_points1, y_points2, y_points3, y_points4]
sub = [221, 222, 223, 224]

for i in range(4):
    plt.subplot(sub[i])
    plt.plot(x_points, y_points[i])
    plt.grid()
    plt.title(type_of_graph[i], fontsize = 15)
    plt.suptitle('All graphs', fontsize = 20)
```

```python
dict_of_data = {}
dict_of_data["x"] = []
dict_of_data["y"] = []
dict_of_data["z"] = []

n = 21
x_points = np.linspace(0, 5, n)
y_points = np.linspace(0, 5, n)
z_points = np.zeros((n, n))
for i in range(len(x_points)):
    for j in range(len(y_points)):
        dict_of_data["x"].append(x_points[i])
        dict_of_data["y"].append(y_points[j])
        z_points[i][j] = 3 * x_points[i] * y_points[j] + x_points[i] - 2 * y_points[j]
        dict_of_data["z"].append(z_points[i][j])
```

```python
import pandas as pd
df = pd.DataFrame(dict_of_data)
df_for_heap = df.pivot("y", "x", "z")

import seaborn as sns
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(df_for_heap, ax=ax)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
axe = Axes3D(fig)
x_points, y_points = np.meshgrid(x_points, y_points)
axe.plot_surface(x_points, y_points, z_points)
axe.set_xlabel("Ось X")
axe.set_ylabel("Ось Y")
axe.set_zlabel("Значение: $3xy + x - 2y$")
plt.show()
```
