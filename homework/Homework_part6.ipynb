### 6 - Pandas + Scikit-learn

Считываем данные с txt-файла и показываю работу со строками и столбцами

```python
import pandas as pd

df = pd.read_csv("Skin_NonSkin.txt", sep = "\t", header = None)
```

```python
print(len(df))
```

```python
df[::20000]
```

```python
df.iloc[[1000, 1010, 1020, 1030], [1, 2]]
```

Добавляю попарное произведение и разницу модулей

```python
df["00"] = df[0] * df[0]
df["01"] = df[0] * df[1]
df["02"] = df[0] * df[2]
df["11"] = df[1] * df[1]
df["12"] = df[1] * df[2]
df["22"] = df[2] * df[2]
df["abs(0-1)"] = abs(df[0] - df[1])
df["abs(0-2)"] = abs(df[0] * df[2])
df["abs(1-2)"] = abs(df[1] * df[2])
```

```python
df
```

Создаю обучающую выборку и тестирующую:

```python
X_init = df.iloc[:, :3]
X_modify = df.iloc[:, 4:]
Y = df[3].values
```

```python
from sklearn.model_selection import train_test_split

X_init_train, X_init_test, Y_train, Y_test = train_test_split(X_init, Y, test_size=0.3, random_state = 11)
X_modify_train, X_modify_test, Y_train, Y_test = train_test_split(X_modify, Y, test_size=0.3, random_state = 11)
```

_______________________________________________

### KNeighboursClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNC_init = KNeighborsClassifier()
model_KNC_init.fit(X_init_train, Y_train)
predicted_values_init_KNC = model_KNC_init.predict(X_init_test)
```

```python
print(sum(predicted_values_init_KNC == Y_test) / float(len(Y_test)))
```

____________

```python
model_KNC_modify = KNeighborsClassifier()
model_KNC_modify.fit(X_modify_train, Y_train)
predicted_values_modify_KNC = model_KNC_modify.predict(X_modify_test)
```

```python
print(sum(predicted_values_modify_KNC == Y_test) / float(len(Y_test)))
```

Таким образом получаем, что алгоритм KNeighboursClassifier на начальных данных работает немного лучше, чем на модифицированной

_________________________

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_LR_init = LogisticRegression()
model_LR_init.fit(X_init_train, Y_train)
predicted_values_init_LR = model_LR_init.predict(X_init_test)
```

```python
print(sum(predicted_values_init_LR == Y_test) / float(len(Y_test)))
```

____

```python
model_LR_modify = LogisticRegression()
model_LR_modify.fit(X_modify_train, Y_train)
predicted_values_modify = model_LR_modify.predict(X_modify_test)
```

```python
print(sum(predicted_values_modify == Y_test) / float(len(Y_test)))
```

Алгоритм логистической регрессии работает лучше на модифицированных данных, чем на начальной
